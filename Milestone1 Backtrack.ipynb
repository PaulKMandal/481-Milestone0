```python
start = (0,0)
end = (0,6)
```

```python
def bfs(start, end, m):
    o = [(start, None)]
    c = []
    
    while o != []:
        X = o.pop(0)
        print(X)
        if X[0] == end:
            return
        temp = []
        if X[0][0] > 0:
            if m[X[0][0] - 1][X[0][1]] == 0:
                temp.insert(0, ((X[0][0] - 1, X[0][1]), X[0]))
        if X[0][0] < 7:
            if m[X[0][0] + 1][X[0][1]] == 0:
                temp.insert(0, ((X[0][0] + 1, X[0][1]),X[0]))
        if X[0][1] > 0:
            if m[X[0][0]][X[0][1] - 1] == 0:
                temp.insert(0, ((X[0][0], X[0][1] - 1),X[0]))
        if X[0][1] < 7: 
            if m[X[0][0]][X[0][1] + 1] == 0:
                temp.insert(0, ((X[0][0], X[0][1] + 1),X[0]))
        
        c.insert(0, X)
        print(temp)
        for child in temp:
            if (not child[0] in [x[0] for x in c]) and (not child[0] in [x[0] for x in o]):
                o.insert(0, child)
        print(c)
```
